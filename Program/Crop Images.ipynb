```python
from PIL import Image
import os
from imutils import paths
```

```python
def crop_img(file_path, crop_size=(300, 300)):
    output_path = file_path.split(os.sep)
    output_path[0] = output_path[0]+'_cropped_'+str(crop_size[0])
    wood_sp = output_path[-1]
    
    os.makedirs(os.path.sep.join(output_path[:-1]), exist_ok=True)
    
    img = Image.open(file_path)
    width, height = img.size
    x, y = (width//crop_size[0], height//crop_size[1])
    
    for j in range(y):
        for i in range(x):
            cropped_img = img.crop((i*crop_size[0], 
                                    j*crop_size[1], 
                                    (i+1)*crop_size[0], 
                                    (j+1)*crop_size[1]))
            output_path[-1] = wood_sp+'_'+str(i)+'_'+str(j)+'.png'
            cropped_img.save(os.path.sep.join(output_path))
            
    print('-', wood_sp,'- Done')
```

```python
image_paths = list(paths.list_images('japanese_fagaceae_dataset'))
```

```python
for img_path in image_paths:
    crop_img(img_path, crop_size=(100, 100))
```
